```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding = 'unicode_escape')
```

```python
df.shape     # to know how many rows and columns are there
```

```python
############### Now we will clean the data ##################
```

```python
df.info()  # will return all the information for the dataset
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)  ## droping these columns as they contain null values, inplace true makes sure
                                                        ##that the change is also done in permanent table 
```

```python
df.info()
```

```python
pd.isnull(df).sum()  ###  this will return sum for all the colums , how many null values are there in that column
```

```python
df.dropna(inplace = True)
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
#### changing the data type for 'amoount' column 
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
### EXploratory Data Analysis (EDA)
```

```python
## Gender###
df.columns
```

```python
ax = sns.countplot(x= 'Gender' , data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)
bx = sns.barplot(x = 'Gender', y = 'Amount', data= Sales_gen)
for bars in bx.containers:
    bx.bar_label(bars)
```

```python
###  AGE  ###
df.columns
```

```python
ax = sns.countplot(x= 'Age Group', data= df, hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x='Age Group', y = 'Amount', data = sales_age)
```

from above graph we can say, the most of the buyers are from age grooup between 26-35 years 

```python
###  State ###
df.columns
```

```python
sales_state = df.groupby(['State'], as_index=False,)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(5)
sns.set(rc={'figure.figsize':(25, 10)})   ####
sns.barplot(x= 'State', y= 'Orders', data = sales_state)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False,)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(5)
sns.set(rc={'figure.figsize':(10, 10)})   #### to set the cha=rt size to see the labels clearly
sx = sns.barplot(x= 'Marital_Status', y= 'Amount', data = sales_state, hue = 'Gender')
for bars in sx.containers:
    sx.bar_label(bars)
```

from the above graph we can say that the most of the buyers are married women 

```python
sales_states = df.groupby(['Occupation', 'Gender'], as_index=False,)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
fx = sns.barplot(x= 'Occupation', y= 'Amount', data = sales_states, hue = "Gender")
for bars in fx.containers:
    fx.bar_label(bars)
```
